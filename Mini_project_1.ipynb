1.Find all the subsets from a set of numbers whose sum is zero.
Constraint: Subset size must be 5
Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
import itertools

def find_zero_sum_subsets(numbers):
    subsets = []
    combinations = itertools.combinations(numbers, 5)
    for combo in combinations:
        if sum(combo) == 0:
            subsets.append(combo)
    return subsets

# Example usage
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
zero_sum_subsets = find_zero_sum_subsets(numbers)
print(zero_sum_subsets)
```

2.Find all the subsets from a set of numbers whose sum is zero.
Constraint: Subset size must be 3 to 6 only
Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
import itertools

def find_zero_sum_subsets(numbers):
    subsets = []
    for size in range(3, 7):
        combinations = itertools.combinations(numbers, size)
        for combo in combinations:
            if sum(combo) == 0:
                subsets.append(combo)
    return subsets

# Example usage
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
zero_sum_subsets = find_zero_sum_subsets(numbers)
print(zero_sum_subsets)
```
